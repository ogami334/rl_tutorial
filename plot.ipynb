```python
import matplotlib.pyplot as plt
import os
import pandas as pd
plt.style.use('dark_background')
```

```python
index = "220430170357"
directory_name = f'pfrl/results/{index}'
score_path = os.path.join(directory_name, 'scores.txt')
df = pd.read_csv(score_path, sep='\t')
# print(df)

# ()
time_step = df['steps']
plt.plot(time_step, df['mean'], label='mean')
plt.plot(time_step, df['max'], label='max')
plt.title('Score')
plt.ylabel('Score')
plt.xlabel('Timesteps')
plt.legend()
plt.show()

plt.xlabel('timestep')
plt.ylabel('Q_value')
plt.title('Q_value')
plt.plot(time_step, df['average_q'])
plt.show()


plt.xlabel('timestep')
plt.ylabel('Loss')
plt.title('Loss')
plt.plot(time_step, df['average_loss'])
plt.show()

```

```python

import datetime

def cur_time():
    t_delta = datetime.timedelta(hours=9)
    JST = datetime.timezone(t_delta, 'JST')
    now = datetime.datetime.now(JST)
    d = now.strftime('%Y%m%d%H%M%S')
    return d[2:]

print(cur_time())
```
