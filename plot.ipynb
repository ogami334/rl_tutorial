```python
import matplotlib.pyplot as plt
import os
import pandas as pd
plt.style.use('dark_background')
```

```python
directory_name = 'result'
score_path = os.path.join(directory_name, 'scores.txt')
df = pd.read_csv(score_path, sep='\t')
# print(df)

# ()
time_step = df['steps']
plt.plot(time_step, df['mean'], label='mean')
plt.plot(time_step, df['max'], label='max')
plt.title('Score')
plt.ylabel('Score')
plt.xlabel('Timesteps')
plt.legend()
plt.show()

plt.xlabel('timestep')
plt.ylabel('Q_value')
plt.title('Q_value')
plt.plot(time_step, df['average_q'])
plt.show()


plt.xlabel('timestep')
plt.ylabel('Loss')
plt.title('Loss')
plt.plot(time_step, df['average_loss'])
plt.show()

```
